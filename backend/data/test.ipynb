```python
import pandas as pd

# Load the dataset
data_path = "/Users/imeag/Documents/udea/trabajoDeGrado/MLOps/backend/data/processed/Data_integration_ic_neuroHarmonize_G1.feather"
df = pd.read_feather(data_path)

# Display the dataset
df
```

```python
# Method 1: Set pandas options to display all columns
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print("All column names:")
print(df.columns.tolist())

print(f"\nTotal number of columns: {len(df.columns)}")

# Method 2: Print each column name on a separate line
print("\nColumn names (one per line):")
for i, col in enumerate(df.columns):
    print(f"{i+1:3d}: {col}")

# Method 3: Group columns by type (power, crossfreq, etc.)
print("\nColumns grouped by type:")
power_cols = [col for col in df.columns if col.startswith('power_')]
crossfreq_cols = [col for col in df.columns if col.startswith('crossfreq_')]

print(f"\nPower columns ({len(power_cols)}):")
for col in power_cols:
    print(f"  {col}")

print(f"\nCrossfreq columns ({len(crossfreq_cols)}):")
for col in crossfreq_cols:
    print(f"  {col}")
```
