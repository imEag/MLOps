```python
# Compare columns between df (original) and df_new (new dataset)
from IPython.display import display
import pandas as pd

# Ensure dataframes exist
if 'df' not in globals() or 'df_new' not in globals():
    raise NameError("Please run the cells that load `df` and `df_new` first.")

cols_df = set(df.columns)
cols_df_new = set(df_new.columns)

only_in_df = sorted(cols_df - cols_df_new)
only_in_df_new = sorted(cols_df_new - cols_df)
common = sorted(cols_df & cols_df_new)

print("Column comparison summary:")
print(f"- Total columns in df: {len(cols_df)}")
print(f"- Total columns in df_new: {len(cols_df_new)}")
print(f"- Columns in both: {len(common)}")
print(f"- Columns only in df: {len(only_in_df)}")
print(f"- Columns only in df_new: {len(only_in_df_new)}")

# Optional: show lists (can be long)
print("\nOnly in df:")
for c in only_in_df:
    print(f"  - {c}")

print("\nOnly in df_new:")
for c in only_in_df_new:
    print(f"  - {c}")

# Tabular summary view
union_cols = sorted(cols_df | cols_df_new)
summary_df = pd.DataFrame({'column': union_cols})
summary_df['in_df'] = summary_df['column'].isin(cols_df)
summary_df['in_df_new'] = summary_df['column'].isin(cols_df_new)
summary_df['status'] = summary_df.apply(
    lambda r: 'both' if r['in_df'] and r['in_df_new'] else ('only_df' if r['in_df'] else 'only_df_new'),
    axis=1,
)

print("\nTabular summary (first 50 rows):")
display(summary_df.head(50))

# Also return the full summary as the cell output
summary_df
```
